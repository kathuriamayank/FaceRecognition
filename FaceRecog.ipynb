```python
#Creating The Function Now
def get_images_labels(path):
    ctr=0
    images=[]
    labels=[]
    images_paths=[]
    
    for ix in os.listdir(path):
        if not ix.endswith(".sad"):
            images_paths.append(ix)
    
    
    #Storing Labels Now
    
    for ix in images_paths:
        labels.append(int(ix.split(".")[0].replace("subject","")))
        #Now Add Faces To Image By Making Use of Haar Cascade
        image_pil = Image.open(os.path.join(path,ix)).convert('L')
        img = np.array(image_pil, 'uint8')
        #img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)


        
        face=facecascade.detectMultiScale(img)
        for (x,y,w,h) in face:
            roi=np.array(img[y:y+h,x:x+w])
            images.append(roi)
            ctr+=1
            cv2.imshow("Image",roi)
            cv2.waitKey(30)
  #  labels=np.array(labels)
   # images=np.array(images)
    
    return images,labels,ctr


    
```

```python
images,labels,ctr=get_images_labels(path)
cv2.destroyAllWindows()
print ctr
print len(images)
print (labels)

```

```python
labels=np.array(labels)
recognizer.train(images,labels)
```

```python
#Testing and Prediction
#it will contain 
images_paths=[]
for ix in os.listdir(path):
    if ix.endswith(".sad"):
        images_paths.append(ix)

        
for ix in images_paths:
    image_pil = Image.open(os.path.join(path,ix)).convert('L')
    img = np.array(image_pil, 'uint8')
    face=facecascade.detectMultiScale(img)
    actual_label=int(ix.split(".")[0].replace("subject",""))
    for (x,y,w,h) in face:
        roi=np.array(img[y:y+h,x:x+w])
        predict_label,confidence=recognizer.predict(roi)
        #print "Actual label is ",actual_label,"Predicted Label is ",predict_label
        if actual_label==predict_label:
            print "The image is predicted correctly for the label ",actual_label,"and its confidence is ",confidence
            cv2.imshow("image",roi)
        else:
            print "The image is predeicted wrong for the label",actual_label
            cv2.imshow("image",roi)

    cv2.waitKey(1000)

        



```
